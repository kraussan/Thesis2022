```python
import numpy as np
import pandas as pd
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
df_1 = pd.read_csv("/content/drive/My Drive/data_1.csv",sep =";", encoding = "cp1251", error_bad_lines=False)
df_2 = pd.read_csv("/content/drive/My Drive/data_2.csv",sep = ";", encoding = "cp1251", error_bad_lines=False)
```

```python
df_all = pd.concat([df_1, df_2])
```

```python
df = df_all
df.drop_duplicates()
```

```python
df['Группа4']
```

```python
df.info()
```

```python
df.drop(['ПричинаОтмены', 'МагазинЗаказа', 'ПВЗ_код'], axis='columns', inplace=True)
```

```python
df = df.dropna(axis = 0, thresh = 30)
```

```python
numeric_columns = ['СуммаЗаказаНаСайте', 'Цена', 'Маржа', 'СуммаУслуг', 'СуммаСтроки']
for column in numeric_columns:
    df[column] = df[column].replace(r'\s+','',regex=True)
    df[column] = df[column].replace(',','.',regex=True)
    df[column] = df[column].apply(pd.to_numeric)
```

```python
df_for_nn = df[['НомерЗаказаНаСайте', 'ID_SKU', 'Группа4', 'Количество', 'ДатаДоставки', 'Цена', 'Телефон_new', 'ГородМагазина']]
```

```python
df_for_nn = df_for_nn.rename({'НомерЗаказаНаСайте': 'InvoiceNo', 'ID_SKU': 'StockCode', 'Группа4': 'Description', 'Количество': 'Quantity', 'ДатаДоставки': 'InvoiceDate', 'Цена': 'UnitPrice','Телефон_new':'Customer ID', 'ГородМагазина':'Country'}, axis = 1)
```

```python
df_for_nn.to_csv('data.csv')
```
